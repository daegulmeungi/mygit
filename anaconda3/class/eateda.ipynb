```python
import pandas as pd
data=pd.read_csv("school_health_2022.csv")
data=data[data['라면'].notna() & data['음료수'].notna() & data['패스트푸드'].notna() & data['육류'].notna() & data['우유유제품'].notna() & data['과일'].notna() & data['채소김치제외'].notna() & data['아침식사'].notna()]
```

```python
import matplotlib.pyplot as plt                      
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'

plt.figure(figsize=(30, 7))
plt.subplot(1, 4, 1)
sns.histplot(data=data, x='라면', discrete=True, kde=False, bins=4)
plt.xticks([1, 2, 3, 4])
plt.title('라면 섭취량')

plt.subplot(1, 4, 2)
sns.histplot(data=data, x='음료수', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('음료수 섭취량')

plt.subplot(1, 4, 3)
sns.histplot(data=data, x='패스트푸드', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('패스트푸드 섭취량')

plt.subplot(1, 4, 4)
sns.histplot(data=data, x='육류', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('육류 섭취량')
```

```python
print("평균:", data['BMI'].mean())
print("중앙값:", data['BMI'].median())
print("분산:", data['BMI'].var())
print("표준편차:", data['BMI'].std())
print("최소값:", data['BMI'].min())
print("최대값:", data['BMI'].max())
```

```python
from scipy.stats import shapiro
from scipy.stats import kstest, norm
import statistics

#정규성 띄는지 확인
# shapiro(data['BMI']) shapiro test를 실시하는 경우 표본의 수가 18000개 가량 되기 때문에 민감하게 반응할 수 있기 때문에 다른 방법을 추천한다.
print(kstest(data['BMI'], norm.rvs(loc=statistics.mean(data['BMI']), scale=statistics.variance(data['BMI']), size=18524)))
```
